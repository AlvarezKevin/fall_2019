```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_html('https://en.wikipedia.org/wiki/List_of_countries_by_median_age')[0]
continents = pd.read_csv('data/continents.csv')
new_df = df.merge(continents, how='inner', left_on='Country/Territory', right_on='Country(or dependent territory)')
new_df.head()
```

## Examples of Different Visualizations Available in Matplotlib

Some useful, but not exhaustive visualizations available in Matplotlib 

### Histogram

"The standard way to show a statistical distribution - keep the gaps between columns small to highlight the 'shape' of the data."

```python
plt.hist(new_df['Average(Years)'])
```

### Bar Chart

Show the counts or values of different discrete values.

Note that if we try to make a bar chart of every country in the dataset it will not be useful because there are hundreds of countries. For your project, think of what metrics will be useful and easily communicated through a graph.

```python
df.plot.bar(x='Country/Territory', y='Average(Years)')
```

```python
df.plot.barh(x='Country/Territory', y='Average(Years)')
```

As an example, we can take the first five countries and make a bar plot of them.

```python
df.head().plot.bar(x='Country/Territory', y='Average(Years)')
```

```python
df.head().plot.bar(x='Country/Territory', y='Average(Years)', legend=False)
```

We can also sort the values for our graph and plot them.

```python
df.head().sort_values('Average(Years)', ascending=False).plot.bar(x='Country/Territory', y='Average(Years)', legend=False)
```

We can also plot multiple values in a bar graph, not just one.

```python
df.head().plot.bar(x='Country/Territory', y=['Average(Years)', 'Male(Years)', 'Female (Years)'])
```

Or flip the bar chart.

```python
df.head().plot.barh(x='Country/Territory', y=['Average(Years)', 'Male(Years)', 'Female (Years)'])
```

### Scatterplots

We will cover correlations next week, but here is how you make a scatterplot if you need it (to show the relationship between two variables)

```python
plt.scatter(df['Male(Years)'], df['Female (Years)'])
```

The pairplot is a good way to examine the relationships between several variables in one image.

```python
sns.pairplot(df[['Average(Years)', 'Male(Years)', 'Female (Years)']])
```

### Line Graphs
Line graphs are good to show values over time. Of course, you can plot multiple line graphs in one image.

```python
plt.plot(df['Average(Years)'])
```

```python
plt.plot(df['Average(Years)'])
plt.plot(df['Male(Years)'])
plt.plot(df['Female (Years)'])
plt.legend()
```

### Heatmap

Heatmaps are good for showing the relative values between two categorical variables.

The best way to do it in a dataframe is similar to below, where you groupby a categorical value and get the sum for the appropriate numerical values.

```python
Index= ['aaa', 'bbb', 'ccc', 'ddd', 'eee']
Cols = ['A', 'B', 'C', 'D']
df = pd.DataFrame(abs(np.random.randn(5, 4)), index=Index, columns=Cols)

sns.heatmap(df, annot=True)
```
