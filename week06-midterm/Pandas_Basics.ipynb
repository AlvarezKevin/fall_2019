## Cleaning Data

```python
cities_by_population = pd.read_html('https://en.wikipedia.org/wiki/List_of_cities_proper_by_population')[1]
cities_by_population.head()
```

```python
#Here I'll remove the 'Image column' as all of its values are null
cities_by_population.pop('Image')
```

```python
cities_by_population.head()
```

### Null Values

See if any of the columns you'll be working with are null. There are a few ways of dealing with it.

Let's say we want to look at the 'population' column - I'll make a few rows null as an example.

```python
np.random.seed(42)
null_values = np.random.choice(cities_by_population.index, size=5)
print(null_values)
cities_by_population['Population'][null_values] = np.nan
```

```python
#Always check to see if any values are null
cities_by_population.isnull().sum()
```

1) If there are a significant number of null values in a column, it may not be worth considering whatsoever and may be best to disregard that column from your analysis. There isn't a go-to standard on this (that I'm aware of), but if *at least* a third to half of a column is unavailable, you should probably disregard it all together.

Below we can see that only 5% of our values for our column is null, so we can explore other options for dealing with it.

```python
cities_by_population.isnull().sum() / len(cities_by_population)
```

One way to address this is to drop rows with null columns. all together. The 'Drop NA' function will drop all rows with *any* missing values

```python
cities_by_population.dropna()
```

Above we'll see that the indexes with null values have been removed.

You can also drop all rows that have *all* missing values (in our case that won't remove anything).

```python
cities_by_population.dropna(how='all')
```

You could also fill the missing rows with the mean value of that column. You could also fill it with the mode, median, or any value that you choose.

```python
cities_by_population['Population'].fillna(10000)
```

```python
cities_by_population['Population'].fillna(cities_by_population['Population'].mean())
```

### Column Types

```python
cities_by_population.dtypes
```

The numerical column we're interested in - **Population** is listed as an 'object'. Numerical columns need to be either an integer or float for us to work with them.

```python
cities_by_population['Population'].head()
```

Specifically our column has two issues - there are commas in it and there are the sources in parenthesis. Below is a way to address this. Knowing exactly what's going on isn't the most important now - feel free to Google or ask me for specific issues with ingesting data. Just know that if you cannot perform numerical operations on your column, such as finding the mean, you need to see why it's not rendering as a number and convert it.

```python
cities_by_population['Population'] = cities_by_population['Population'].str.split('[').str[0].str.replace(',','')
```

Confirm that any numerical columns you want to work with are listed as integers or floats. To change a column to an integer or a float, you can try:

```python
cities_by_population['Population'] = cities_by_population['Population'].astype('float')
```

```python
cities_by_population['Population'] = cities_by_population['Population'].fillna(cities_by_population['Population'].mean())
```
